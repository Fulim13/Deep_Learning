# Exercise 1 (Single Input Output Neural Network):

Given weight = `0.1` and number_of_toes = `[8.5, 9.5, 10, 9]`

Create a single input output neural network that take the first element of `number_of_toes` as `input` with `weight` to create a prediction `pred`



````python
weight = 0.1
number_of_toes = [8.5, 9.5, 10, 9]

input = number_of_toes[0]

def neural_network(input,weight):
    return input * weight

pred = neural_network(input,weight)
print(pred)
````

# Exercise 2 (Multiple Input Single Output Neural Network):

Given `weights = [0.1, 0.2, 0]`
and 
`toes =  [8.5, 9.5, 9.9, 9.0]`
`wlrec = [0.65, 0.8, 0.8, 0.9]`
`nfans = [1.2, 1.3, 0.5, 1.0]`

Create a multiple inputs single output neural network that take the first element of `toes`, `wlrec`, and `nfans` as `inputs` with `weights` to create a prediction `pred`



````python
weights = [0.1, 0.2, 0]

toes =  [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

inputs = [toes[0],wlrec[0],nfans[0]]

def w_sum(a,b):
    assert(len(a) == len(b))
    output = 0
    for i in range(len(a)):
        output += a[i] * b[i]
    
    return output

def neural_network(inputs, weights):
    return w_sum(inputs,weights)

pred = neural_network(inputs,weights)
print(pred)
````

# Exercise 3 (Multiple Input Single Output Neural Network With Numpy):

Given `weights = [0.1, 0.2, 0]`
and 
`toes =  [8.5, 9.5, 9.9, 9.0]`
`wlrec = [0.65, 0.8, 0.8, 0.9]`
`nfans = [1.2, 1.3, 0.5, 1.0]`

Create a multiple inputs single output neural network that take the first element of `toes`, `wlrec`, and `nfans` as `inputs` with `weights` to create a prediction `pred`



````python
import numpy as np

weights = np.array([0.1, 0.2, 0])

toes =  np.array([8.5, 9.5, 9.9, 9.0])
wlrec = np.array([0.65, 0.8, 0.8, 0.9])
nfans = np.array([1.2, 1.3, 0.5, 1.0])

inputs = np.array([toes[0], wlrec[0], nfans[0]])

def neural_network(inputs,weights):
    return inputs.dot(weights)

pred = neural_network(inputs,weights)
print(pred)
````

# Exercise 4 (Single Input Multiple Output Neural Network):

Given `weights = [0.3, 0.2, 0.9] ` and wlrec = `[0.65, 0.8, 0.8, 0.9]`, create a single input multiple output neural network that take the first element of `wlrec` as `input` with `weights` to create a prediction `pred`



````python
weights = [0.3, 0.2, 0.9]

wlrec = [0.65, 0.8, 0.8, 0.9]
input = wlrec[0]

def ele_mul(number,vector):
    output = [0, 0, 0]
    assert(len(output) == len(vector))
    for i in range(len(vector)):
        output[i] = number * vector[i]
    return output

def neural_network(input,weights):
    return ele_mul(input,weights)

pred = neural_network(input,weights)
print(pred)
````

# Exercise 6 (Multiple Input Multiple Output Neural Network):

Given 
```
            #toes %win #fans
weights = [[0.1, 0.1, -0.3], # hurt?
           [0.1, 0.2, 0.0], # win?
           [0.0, 1.3, 0.1]] # sad?
```

and 
```
toes =  [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]
```

Create a multiple inputs multiple output neural network that take the first element of `toes`, `wlrec`, and `nfans` as `inputs` with `weights` to create a prediction `pred`



````python
weights = [ [0.1, 0.1, -0.3], 
            [0.1, 0.2, 0.0], 
            [0.0, 1.3, 0.1] ]

toes =  [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65,0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

inputs = [toes[0],wlrec[0], nfans[0]]

def w_sum(a,b):
    assert(len(a) == len(b))
    output = 0
    for i in range(len(a)):
        output += a[i] * b[i]
    return output

def vect_mat_mul(vector, matrix):
    assert(len(vector) == len(matrix))
    output = [0,0,0]
    for i in range(len(matrix)):
        output[i] = w_sum(vector,matrix[i])

    return output

def neural_network(inputs,weights):
    return vect_mat_mul(inputs,weights)

pred = neural_network(inputs,weights)
print(pred)
````

````python
            #toes %win #fans
ih_wgt = [ [0.1, 0.2, -0.1], #hid[0]
           [-0.1,0.1, 0.9], #hid[1]
           [0.1, 0.4, 0.1] ] #hid[2]

           #hid[0] hid[1] hid[2]
hp_wgt = [ [0.3, 1.1, -0.3], #hurt?
           [0.1, 0.2, 0.0], #win?
           [0.0, 1.3, 0.1] ] #sad?

weights = [ih_wgt, hp_wgt]

toes =  [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65,0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]


inputs = [toes[0],wlrec[0],nfans[0]]

def w_sum(a,b):
    assert(len(a) == len(b))
    output = 0
    for i in range(len(a)):
        output += a[i] * b[i]

    return output


def vect_mat_mul(vector, matrix):
    assert(len(vector) == len(matrix))
    output = [0, 0, 0]
    for i in range(len(matrix)):
        output[i] = w_sum(vector,matrix[i])

    return output

def neural_network(inputs, weights):
    hid = vect_mat_mul(inputs, weights[0])
    pred = vect_mat_mul(hid, weights[1])
    return pred

pred = neural_network(inputs, weights)
print(pred)
````
